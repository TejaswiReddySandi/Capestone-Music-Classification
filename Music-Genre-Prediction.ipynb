```python
# preparing data
traindata = train.copy()
```

```python
traindata.shape
# Artist Name            9149
# Track Name            15129
```

```python
traindata.columns
```

```python
# Frequency encoding for columns with large unique values

# For 'Artist Name'
artist_name_counts = traindata['Artist Name'].value_counts().to_dict()
traindata['Artist Name Encoded'] = traindata['Artist Name'].map(artist_name_counts)

# For 'Track Name'
track_name_counts = traindata['Track Name'].value_counts().to_dict()
traindata['Track Name Encoded'] = traindata['Track Name'].map(track_name_counts)

# Drop the original columns if needed

#traindata = traindata.drop(['Artist Name', 'Track Name'], axis=1)
traindata
```

```python
train_x = traindata.drop(['Artist Name','Track Name','Class','Class_category'], axis=1)
train_y = traindata['Class']
train_x.columns
```

```python
train_y.isnull().sum()
```

```python
train_x = train_x.drop(['Artist Name Encoded', 'Track Name Encoded'], axis=1)
train_x
```

```python
# Standardising the data for the SMOTE technique
from sklearn.preprocessing import StandardScaler

# Assuming train_x is your feature matrix with numerical columns
scaler = StandardScaler()
x_standardized = scaler.fit_transform(train_x)
```

```python
#resampling using resample

from sklearn.utils import resample
```

```python
# Initialize SMOTE
# 'auto' balances all classes
smote = SMOTE(sampling_strategy='auto', random_state=42)
```

```python
# applying SMOTE
x_resampled, y_resampled = smote.fit_resample(x_standardized, train_y)
```
